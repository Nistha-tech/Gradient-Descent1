```python
import numpy as np
import pandas as pd
from sklearn import preprocessing
scaler=preprocessing.StandardScaler()
```

```python
data=np.loadtxt(r"C:\Users\hp\Downloads\0000000000002419_training_ccpp_x_y_train.csv",delimiter=",")
df=pd.DataFrame(data)
y=data[:,4]
data=np.delete(data,4,1)
c=np.ones((7176,1),dtype=float)
scaler.fit(data)
data=scaler.transform(data)
data=np.append(data,c,1)
data.shape
```

```python
def stepgr(data,learning_rate,m):
    m_slope=np.zeros(5)
    M=len(data)
    for i in range(5):
        for j in range(M):
            m_slope[i]+=(-2/M)*(y[j]-(m * data[j,0:5]).sum())*data[j,i]
    new_m=m-learning_rate*m_slope
    return new_m
        
```

```python
def gd(data,learning_rate,num):
    m=np.zeros(5)
    for i in range(num):
        m=stepgr(data,learning_rate,m)
        print(i," cost:",cost(data,m))
    return m
```

```python
def cost(data,m):
    total_cost=0
    M=len(data)
    for i in range(M):
        X=data[i]
        Y=y[i]
        total_cost+=(1/M)*((Y-(m*X).sum())**2)
    return total_cost
```
